#คำใบ้2

I was hired by a woman with a lot of money. I don't know her name but I know she's around 5'5" (65") or 5'7" (67"). She has red hair and she drives a Tesla Model S. I know that she attended the SQL Symphony Concert 3 times in December 2017.

 กรองตามเงื่อนไข
   - female
   - height 65–67 นิ้ว
   - hair_color = red
    - car_make = Tesla
    - car_model = Model S

```python
event_file = "facebook_event_checkin.csv"
license_file = "drivers_license.csv"

event_data = pd.read_csv(event_file)
license_data = pd.read_csv(license_file)
```

```python
# 1) หา event_id ของ SQL Symphony Concert
sql_concerts = event_data[event_data['event_name'].str.contains("SQL Symphony Concert", case=False, na=False)]
sql_concerts.head()
```

```python
# 2) หา person_id ที่ไป SQL Symphony Concert อย่างน้อย 3 ครั้งในเดือน Dec 2017
sql_concerts_dec2017 = sql_concerts[sql_concerts['date'].between(20171201, 20171231)]

# นับจำนวนครั้งที่เข้าร่วมต่อ person_id
concert_counts = sql_concerts_dec2017.groupby("person_id").size().reset_index(name="count")

# เลือกเฉพาะคนที่ไป >= 3 ครั้ง
suspects_concert = concert_counts[concert_counts["count"] >= 3]
suspects_concert
```
