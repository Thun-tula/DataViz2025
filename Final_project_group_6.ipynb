```python
import pandas as pd
import plotly.express as px

# ---------- 1) เตรียมข้อมูล ----------
graph = merged_df_clear.copy()

graph["Continent"] = "Europe"

def infer_country(region: str) -> str:
    if pd.isna(region):
        return "Unknown"
    s = str(region)
    if "England" in s: return "England"
    if "Wales" in s: return "Wales"
    if "Scotland" in s: return "Scotland"
    if "Northern Ireland" in s: return "Northern Ireland"
    return "United Kingdom"

graph["Country"] = graph["Region"].apply(infer_country)
graph["Region"] = graph["Region"].fillna("Unknown")
graph["LADs"] = graph["LADs"].fillna("Unknown")

# ---------- 2) สร้างตารางสรุป ----------
sun = (
    graph.groupby(["Continent","Country","Region","LADs"], dropna=False)
         .size()
         .reset_index(name="Crime_count")
         .sort_values("Crime_count", ascending=False)
)

# ลดจำนวนเมือง (LADs) ที่จะโชว์ — แสดงเฉพาะ 50 เมืองที่มีคดีเยอะที่สุด
TOP_N = 50
sun_filtered = sun.groupby("Region").head(int(TOP_N / 4)).copy()  # เฉลี่ยให้แต่ละ region
if sun_filtered.empty:
    sun_filtered = sun.copy()

# ---------- 3) วาด Sunburst ----------
fig = px.sunburst(
    sun_filtered,
    path=["Continent","Country","Region","LADs"],
    values="Crime_count",
    color="Crime_count",
    color_continuous_scale="YlOrRd",
    color_continuous_midpoint=sun_filtered["Crime_count"].median(),
    maxdepth=3  # จำกัดระดับชั้น (ป้องกัน label ซ้อน)
)

# ---------- 4) ปรับการแสดงผล ----------
fig.update_traces(
    textinfo="label+percent entry",   # แสดงชื่อ + เปอร์เซ็นต์
    insidetextorientation="radial",   # หมุนข้อความให้ตามวงกลม
    hovertemplate="<b>%{label}</b><br>Crimes: %{value:,}<extra></extra>"
)

fig.update_layout(
    title="Crimes by Continent → Country → Region → City",
    title_font=dict(size=18),
    font=dict(size=11),
    margin=dict(t=50, l=10, r=10, b=10),
    coloraxis_colorbar=dict(title="Crimes"),
    height=700,
    width=700
)

fig.show()

# ---------- 5) บันทึกเป็น HTML ----------
fig.write_html("sunburst_crimes_uk.html")
print("✅ Saved:", "sunburst_crimes_uk.html")
```

```python
# ===================== Criminal Justice Pyramid (Success vs Fail by City) =====================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

graph = merged_df_clear  # ใช้ตัวที่คุณทำความสะอาดแล้ว


# ---------- 1) ตั้งค่าคอลัมน์ ----------
city_col = "LADs"         # <<<<<< สำคัญ: ใช้ชื่อนี้แทน 'City'
region_col = "Region"             # ถ้าจะฟิลเตอร์ราย Region ได้
USE_PERCENT = False                  # เปลี่ยนเป็น True ถ้าอยากดูเปอร์เซ็นต์
TOP_CITIES = 12
RANK_BY = "total"                    # "total" หรือ "fail_rate"
REGION_FILTER = None                 # เช่น "West Midlands" หรือ None ไม่ฟิลเตอร์

# ---------- 2) เตรียม outcome (Outcome type > Last outcome category) ----------
outcome_series = graph.get('Outcome type').fillna(graph.get('Last outcome category')) # Corrected to use 'graph'

def map_outcome(x: str) -> str:
    if not isinstance(x, str):
        return "Other"
    s = x.strip().lower()

    # กลุ่มสำเร็จ
    success_kw = [
        "charged", "charge", "court result", "summons",
        "conditional caution", "caution",
        "penalty notice",
        "community resolution", "local resolution",
        "drugs possession warning",
        "fine", "fined",
        "prison", "community sentence"
    ]
    # กลุ่มไม่สำเร็จ
    fail_kw = [
        "no suspect identified",
        "unable to prosecute",
        "status update unavailable",
        "further investigation is not in the public interest",
        "evidential difficulties",
        "not in the public interest",
    ]
    # กลุ่มที่ยังดำเนินการ (ให้นับเป็น Other จะได้ไม่ทำให้ success/fail บิด)
    other_kw = [
        "under investigation", "awaiting court result", "under investigation"
    ]

    if any(k in s for k in success_kw):
        return "Success"
    if any(k in s for k in fail_kw):
        return "Fail"
    if any(k in s for k in other_kw):
        return "Other"
    return "Other"

graph = graph.copy() # Corrected to use 'graph'
graph['OutcomeBin'] = outcome_series.apply(map_outcome) # Corrected to use 'graph'

# ---------- 3) เลือกเฉพาะ Success/Fail และคอลัมน์เมือง ----------
if city_col not in graph.columns: # Corrected to use 'graph'
    raise KeyError(f"ไม่พบคอลัมน์เมือง '{city_col}' ใน DataFrame")

if REGION_FILTER and region_col in graph.columns: # Corrected to use 'graph'
    graph = graph[graph[region_col] == REGION_FILTER] # Corrected to use 'graph'

graph[city_col] = graph[city_col].fillna('Unknown') # Corrected to use 'graph'

df_pyr = graph[graph['OutcomeBin'].isin(['Success', 'Fail'])] # Corrected to use 'graph'
pv = df_pyr.pivot_table(index=city_col, columns='OutcomeBin', aggfunc='size', fill_value=0)

# เติมให้ครบสองฝั่ง
for col in ['Success', 'Fail']:
    if col not in pv.columns:
        pv[col] = 0

pv['Total'] = pv['Success'] + pv['Fail']
pv['Fail_rate'] = pv['Fail'] / pv['Total'].replace(0, np.nan)

# เลือกเกณฑ์จัดอันดับ
if RANK_BY == "fail_rate":
    pv = pv.sort_values(['Fail_rate', 'Total'], ascending=[False, False])
else:
    pv = pv.sort_values('Total', ascending=False)

pv_top = pv.head(TOP_CITIES).copy()

# ---------- 4) เปลี่ยนเป็นเปอร์เซ็นต์ถ้าต้องการ ----------
metric_label = "Number of cases"
if USE_PERCENT:
    pv_top[['Success', 'Fail']] = pv_top[['Success','Fail']].div(pv_top['Total'].replace(0, np.nan), axis=0) * 100
    pv_top[['Success', 'Fail']] = pv_top[['Success', 'Fail']].fillna(0)
    metric_label = "Percentage (%)"

# ---------- 5) เตรียมค่าพล็อต ----------
left_vals  = -pv_top['Success'].values
right_vals =  pv_top['Fail'].values
cities = pv_top.index.tolist()

# ---------- 6) วาดกราฟพีระมิด ----------
fig, ax = plt.subplots(figsize=(12, 7))
y = np.arange(len(cities))

bars_left  = ax.barh(y, left_vals,  align='center', label='Success')
bars_right = ax.barh(y, right_vals, align='center', label='Fail')

ax.set_yticks(y)
ax.set_yticklabels(cities)
ax.set_xlabel(metric_label)
title_rank = "by Total" if RANK_BY=="total" else "by Fail Rate"
subtitle = f"Top {TOP_CITIES} cities {title_rank}" + (f" | Region: {REGION_FILTER}" if REGION_FILTER else "")
ax.set_title(f"Criminal Justice Pyramid (Success vs Fail) — {subtitle}")

# ให้แกน X แสดงเป็นค่าบวก
def fmt_abs(x, pos): return f"{abs(int(x))}"
ax.xaxis.set_major_formatter(plt.FuncFormatter(fmt_abs))

# (ทางเลือก) ใส่ตัวเลขบนปลายแท่ง
for b in list(bars_left) + list(bars_right):
    w = b.get_width()
    if w == 0:
        continue
    ax.text(b.get_x() + w + (15 if not USE_PERCENT else 1),
            b.get_y() + b.get_height()/2,
            f"{abs(w):.0f}" if not USE_PERCENT else f"{abs(w):.1f}%",
            va='center', ha='left' if w>0 else 'right', fontsize=9)

ax.legend(loc='upper right')
ax.grid(axis='x', alpha=0.2)
plt.tight_layout()
plt.show()

# ---------- 7) Preview ตาราง ----------
display_cols = ['Success', 'Fail', 'Total', 'Fail_rate']
if USE_PERCENT:
    display_cols = ['Success', 'Fail', 'Total', 'Fail_rate']
print("\nPreview:")
print(pv_top[display_cols].assign(Fail_rate=(pv_top['Fail']/pv_top['Total']).round(3)).round(1 if USE_PERCENT else 0))
```

```python
import pandas as pd
import numpy as np
import re

# -------------------------------
# 1. ฟังก์ชันช่วย: หา column ตามชื่อ (ไม่แคร์ case / space / underscore)
# -------------------------------
def find_col(df, patterns):
    cols = []
    for col in df.columns:
        c = col.lower().replace(" ", "_")
        for p in patterns:
            if re.fullmatch(p.lower().replace(" ", "_"), c):
                cols.append(col)
    return cols

def match_col(df, patterns):
    cols = find_col(df, patterns)
    return cols[0] if cols else None

# -------------------------------
# 2. ลบคอลัมน์ที่ไม่ต้องการ
# -------------------------------
drop_cols = find_col(sns_df1, ["part_of_a_policing_operation", "policing_operation"])
sns_df1 = sns_df1.drop(columns=drop_cols, errors='ignore')

# -------------------------------
# 3. ลบแถวที่ legislation และ object_of_search ว่างพร้อมกัน
# -------------------------------
leg_col = match_col(sns_df1, ["legislation"])
obj_col = match_col(sns_df1, ["object_of_search", "object of search", "Object_of_search"])

if leg_col and obj_col:
    sns_df1 = sns_df1.dropna(subset=[leg_col, obj_col], how='all')

# -------------------------------
# 4. ลบแถวที่พิกัดว่างหรือผิดปกติ
# -------------------------------
lat_col = match_col(sns_df1, ["latitude"])
lon_col = match_col(sns_df1, ["longitude"])

if lat_col and lon_col:
    sns_df1 = sns_df1.dropna(subset=[lat_col, lon_col])
    sns_df1 = sns_df1[
        (sns_df1[lat_col].between(-90, 90, inclusive="both")) &
        (sns_df1[lon_col].between(-180, 180, inclusive="both")) &
        ~((sns_df1[lat_col] == 0) & (sns_df1[lon_col] == 0))
    ]

# -------------------------------
# 5. เติมค่า "not provided" ยกเว้น Self-defined ethnicity
# -------------------------------
for colname in [
    "Gender", "Age range", "Officer-defined ethnicity", "Outcome linked to object of search"
]:
    c = match_col(sns_df1, [colname])
    if c:
        sns_df1[c] = sns_df1[c].fillna("not provided")

sns_df1['Age range'] = sns_df1['Age range'].str.replace('-', '~')
# สร้าง mapping สำหรับจัดกลุ่มเชื้อชาติ
mapping = {
    'White': [
        'White - English/Welsh/Scottish/Northern Irish/British',
        'White - Any other White background',
        'White - Irish',
        'White - Gypsy or Irish Traveller'
    ],
    'Black': [
        'Black/African/Caribbean/Black British - African',
        'Black/African/Caribbean/Black British - Caribbean',
        'Black/African/Caribbean/Black British - Any other Black/African/Caribbean background'
    ],
    'Asian': [
        'Asian/Asian British - Indian',
        'Asian/Asian British - Pakistani',
        'Asian/Asian British - Any other Asian background',
        'Asian/Asian British - Bangladeshi',
        'Asian/Asian British - Chinese'
    ],
    'Mixed': [
        'Mixed/Multiple ethnic groups - White and Asian',
        'Mixed/Multiple ethnic groups - White and Black Caribbean',
        'Mixed/Multiple ethnic groups - Any other Mixed/Multiple ethnic background',
        'Mixed/Multiple ethnic groups - White and Black African'
    ],
    'Other': [
        'Other ethnic group - Any other ethnic group',
        'Other ethnic group - Not stated'
        'Other ethnic group - Arab'
    ]
}

# ฟังก์ชันแม็ปค่าเชื้อชาติให้เข้ากลุ่มหลัก
def map_ethnicity(val):
    if pd.isna(val):
        return 'not provided'
    for group, values in mapping.items():
        if val in values:
            return group
    return 'Other'  # กรณีไม่เจอใน mapping ใดเลย

# Rename the column first
sns_df1 = sns_df1.rename(columns={'Self-defined ethnicity': 'ethnicity'})

# Corrected: Apply the function to the 'ethnicity' column directly
sns_df1['ethnicity_group'] = sns_df1['ethnicity'].apply(map_ethnicity)


# -------------------------------
# 8. แตกวันเวลา
# -------------------------------
date_col = match_col(sns_df1, ["date"])
if date_col:
    # ลบคอลัมน์ date/time/year/month/day เดิม
    sns_df1 = sns_df1.drop(columns=[c for c in sns_df1.columns if re.search(r"(^time$|^year$|^month$|^day$)", c, flags=re.IGNORECASE)], errors='ignore')
    # แปลงเป็น datetime

# -------------------------------
# 9. เดา / เติมค่า legislation จาก object_of_search และกลับกัน
# -------------------------------
if leg_col and obj_col:
    # เริ่มจากคู่จริงที่พบ
    valid_pairs = sns_df1[[leg_col, obj_col]].dropna()
    pair_mode = valid_pairs.groupby(obj_col)[leg_col].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    obj_to_leg = pair_mode.to_dict()

    # เติม legislation จาก object_of_search
    sns_df1[leg_col] = sns_df1.apply(
        lambda r: obj_to_leg.get(r[obj_col], r[leg_col]) if pd.isna(r[leg_col]) and not pd.isna(r[obj_col]) else r[leg_col],
        axis=1
    )

    # เติมด้วย heuristic ถ้ายังว่าง
    heuristics = {
        "Controlled drugs": "Misuse of Drugs Act 1971 (section 23)",
        "Offensive weapons": "Police and Criminal Evidence Act 1984 (section 1)",
        "Stolen goods": "Police and Criminal Evidence Act 1984 (section 1)",
        "Articles for use in theft": "Police and Criminal Evidence Act 1984 (section 1)",
        "Firearms": "Firearms Act 1968 (section 47)",
        "Psychoactive substances": "Psychoactive Substances Act 2016 (section 36(2))",
    }
    for key, val in heuristics.items():
        sns_df1.loc[
            sns_df1[leg_col].isna() & sns_df1[obj_col].str.contains(key, case=False, na=False),
            leg_col
        ] = val

    # เติม object_of_search จาก legislation (เช่น s60)
    sns_df1.loc[
        sns_df1[obj_col].isna() & sns_df1[leg_col].str.contains("Criminal Justice and Public Order Act 1994", case=False, na=False),
        obj_col
    ] = "Offensive weapons"

# -------------------------------
# 10. Final sweep: เติม "not provided" ทุกข้อความที่ยังว่าง
# -------------------------------
for c in sns_df1.select_dtypes(include=['object', 'string']).columns:
    sns_df1[c] = sns_df1[c].fillna("not provided").replace("", "not provided")

print(" Data cleaned successfully!")
```

```python
# =========================================================
# Grouped Bar: Average per Day — Weekday vs Weekend by Region
# - Day: Monday..Sunday  -> WeekType (Weekday / Weekend)
# - ค่า = Avg per day (Weekday ÷ 5, Weekend ÷ 2)
# - จัดเรียง Region ตาม Weekday Avg มาก→น้อย
# - เลือก Top-N Region เพื่อลดความแน่นได้
# =========================================================
!pip -q install plotly

import pandas as pd, numpy as np
import plotly.express as px

# ---------- 1) Load ----------
def get_df():
    try:
        # Corrected: Check if filtered_df exists and is not None
        if 'filtered_df' in globals() and filtered_df is not None:
            graph2 = filtered_df.copy()
            print("✅ Using DataFrame: filtered_df")
            return graph2
        else:
            raise NameError("filtered_df is not defined or is None")
    except NameError:
        # ถ้าไม่มี filtered_df ให้ลองอ่านจากไฟล์ที่อัปไว้
        path = "/mnt/data/filtered_data.csv"
        try:
            graph2 = pd.read_csv(path)
            print(f"✅ Loaded from file: {path}")
            return graph2
        except FileNotFoundError:
            print(f"❌ Error: File not found at {path}")
            # Return an empty DataFrame with expected columns to prevent downstream errors
            return pd.DataFrame(columns=['Region', 'Day'])


# Corrected: Use graph2_raw instead of df_raw
graph2_raw = get_df()

# Check if graph2_raw is empty before proceeding
if graph2_raw.empty:
    print("❗ Data loading failed or resulted in an empty DataFrame. Cannot proceed with plotting.")
else:
    # ---------- 2) Resolve columns ----------
    def normkey(s): return "".join(ch for ch in str(s).lower() if ch.isalnum())
    # Corrected: Use graph2_raw instead of df_raw
    cols = {normkey(c): c for c in graph2_raw.columns}

    region_col = next((v for k,v in cols.items() if "region" in k), None)
    day_col    = next((v for k,v in cols.items() if k=="day" or k.endswith("day")), None)
    if not region_col or not day_col:
        # Corrected: Raise an error or print a message if essential columns are missing
        print("❌ Error: Essential columns 'Region' or 'Day' not found in the DataFrame.")
    else:
        # Corrected: Use graph2_raw instead of df_raw and graph2 instead of df
        graph2 = graph2_raw[[region_col, day_col]].copy()
        graph2.columns = ["Region", "Day"]
        graph2["Region"] = graph2["Region"].astype(str).str.strip()
        graph2["Day"]    = graph2["Day"].astype(str).str.strip().str.title()

        # ---------- 3) WeekType ----------
        WEEKDAY = {"Monday","Tuesday","Wednesday","Thursday","Friday"}
        WEEKEND = {"Saturday","Sunday"}

        def to_weektype(d):
            if d in WEEKDAY: return "Weekday"
            if d in WEEKEND: return "Weekend"
            return np.nan

        # Corrected: Use graph2 instead of df
        graph2["WeekType"] = graph2["Day"].apply(to_weektype)
        # Corrected: Use graph2 instead of df
        graph2 = graph2.dropna(subset=["Region","WeekType"])

        # ---------- 4) Count -> Avg per day ----------
        days_per = {"Weekday":5, "Weekend":2}
        # Corrected: Use graph2 instead of df
        g = (graph2.groupby(["Region","WeekType"]).size().reset_index(name="Count"))
        g["AvgPerDay"] = g.apply(lambda r: r["Count"] / days_per[r["WeekType"]], axis=1)

        # เติม 0 ให้ region ที่ขาดบางสัปดาห์
        pivot = g.pivot_table(index="Region", columns="WeekType", values="AvgPerDay", fill_value=0.0)

        # เรียง Region ตาม Weekday Avg มาก→น้อย
        if "Weekday" not in pivot.columns: pivot["Weekday"] = 0.0
        if "Weekend" not in pivot.columns: pivot["Weekend"] = 0.0
        pivot = pivot.sort_values("Weekday", ascending=False)

        # (ทางเลือก) โฟกัสเฉพาะ Top-N region
        TOP_N = None   # เช่น 15; หรือ None = ทั้งหมด
        if TOP_N:
            pivot = pivot.head(TOP_N)

        plot_df = pivot.reset_index().melt(id_vars="Region", value_vars=["Weekday","Weekend"],
                                           var_name="WeekType", value_name="AvgPerDay")

        # ---------- 5) Plot ----------
        fig = px.bar(
            plot_df,
            x="Region", y="AvgPerDay", color="WeekType",
            barmode="group",
            text=plot_df["AvgPerDay"].round(1),
            title="Average Stop & Search per Day — Weekday vs Weekend by Region",
            labels={"AvgPerDay":"Average per Day", "Region":"Region", "WeekType":""},
            category_orders={"Region": plot_df["Region"].tolist()}  # คงลำดับตามการเรียง
        )

        fig.update_traces(textposition="outside", cliponaxis=False)
        fig.update_layout(
            xaxis_tickangle=-35,
            yaxis=dict(zeroline=True, rangemode="tozero"),
            margin=dict(t=70,l=40,r=20,b=120),
            legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5)
        )

        fig.show()

        # (optional) save เป็น HTML
        out_path = "/content/bar_weektype_region_avg.html"
        fig.write_html(out_path)
        print(f"💾 Saved: {out_path}")
```

```python
# Grouped + Stacked Bar: Age range × (Male/Female), stacked by Object of search (Percent)
!pip -q install plotly
import pandas as pd, numpy as np
import plotly.graph_objects as go
import plotly.express as px


graph2 = filtered_df.copy(); print("✅ filtered_df")


# ----- select & clean -----
# Corrected: Include 'Region' in the column selection
graph2 = graph2[["Age range","Gender","Object of search", "Region"]].copy()
# Corrected: Use graph2 instead of df
graph2["Age range"] = graph2["Age range"].astype(str).str.strip().str.title().replace({"Not specified":"Not Specified"})
# Corrected: Use graph2 instead of df
graph2["Gender"]    = graph2["Gender"].astype(str).str.strip().str.title()
# Corrected: Use graph2 instead of df
graph2["Object of search"] = graph2["Object of search"].astype(str).str.strip()
# Corrected: Use graph2 instead of df
graph2 = graph2[graph2["Gender"].isin(["Male","Female"])]

# *** Debugging: Inspect unique values in 'Age range' after replace ***
print("Unique values in 'Age range' after replace:")
print(graph2['Age range'].unique())
# *** End Debugging ***


age_order = ["10-17","18-24","25-34","Over 34","Not Specified"]
# normalize ให้ "Not specified" กลายเป็น "Not Specified"
# Corrected: Update replace mapping to handle '~' and other values
graph2["Age range"] = graph2["Age range"].replace({
    "Not specified": "Not Specified",
    "18~24": "18-24",
    "25~34": "25-34",
    "10~17": "10-17",
    "Over 34": "Over 34",
    "Not Provided": "Not Specified", # Map 'Not Provided' to 'Not Specified'
    "Under 10": "10-17" # Map 'Under 10' to '10-17' or create a new category
})

# Corrected: Use graph2 instead of df
graph2["Age range"] = pd.Categorical(graph2["Age range"], categories=age_order, ordered=True)

# ตัดแถวที่คีย์หลักว่าง (อย่างน้อยต้องมี Gender กับ Age range)
# Corrected: Use graph2 instead of df
graph2 = graph2.dropna(subset=["Gender", "Age range", "Region"])

# ----- aggregate & % within (Age range + Gender) -----
# Corrected: Use graph2 instead of df and add observed=False
g = (graph2.groupby(["Age range","Gender","Object of search"], dropna=False, observed=False).size()
       .reset_index(name="Count"))
g["Pct"] = g.groupby(["Age range","Gender"], observed=False)["Count"].transform(lambda s: s/s.sum()*100).fillna(0)

# (optional) limit objects to Top-N overall for readability
TOP_OBJECTS = None   # e.g., 10
if TOP_OBJECTS:
    topO = (g.groupby("Object of search")["Count"].sum()
              .sort_values(ascending=False).head(TOP_OBJECTS).index)
    g = g[g["Object of search"].isin(topO)]

# ensure every Age×Gender has all objects
all_age = [a for a in age_order if a in g["Age range"].unique()]
all_obj = sorted(g["Object of search"].unique())
idx = pd.MultiIndex.from_product([all_age, ["Male","Female"], all_obj],
                                 names=["Age range","Gender","Object of search"])
g = g.set_index(["Age range","Gender","Object of search"]).reindex(idx, fill_value=0).reset_index()

# ----- plot (two side-by-side stacked bars per age) -----
palette = px.colors.qualitative.Set2
color_map = {obj: palette[i % len(palette)] for i, obj in enumerate(all_obj)}

fig = go.Figure()
for obj in all_obj:
    # Male stack
    sub = g[(g["Gender"]=="Male") & (g["Object of search"]==obj)]
    fig.add_trace(go.Bar(
        x=sub["Age range"], y=sub["Pct"],
        name=obj, marker_color=color_map[obj],
        offsetgroup="Male",  # <- ทำให้แท่งเพศชายอยู่ก้อนเดียวกัน
        legendgroup=obj, showlegend=True,
        hovertemplate="<b>%{x}</b><br>Male — "+obj+"<br>%{y:.1f}%<extra></extra>"
    ))
    # Female stack
    sub = g[(g["Gender"]=="Female") & (g["Object of search"]==obj)]
    fig.add_trace(go.Bar(
        x=sub["Age range"], y=sub["Pct"],
        name=obj, marker_color=color_map[obj],
        offsetgroup="Female",  # <- ทำให้แท่งเพศหญิงขยับไปข้าง ๆ
        legendgroup=obj, showlegend=False,
        hovertemplate="<b>%{x}</b><br>Female — "+obj+"<br>%{y:.1f}%<extra></extra>"
    ))

fig.update_layout(
    barmode="relative",  # <- ซ้อนภายในแต่ละ offsetgroup
    title="Object of Search by Age Range — Male vs Female (% within group)",
    xaxis_title="Age range", yaxis_title="Percent",
    yaxis=dict(ticksuffix="%", rangemode="tozero"),
    legend=dict(orientation="h", yanchor="bottom", y=-0.18, xanchor="center", x=0.5),
    margin=dict(t=70,l=40,r=20,b=120)
)
fig.update_traces(width=0.45)
fig.show()
```
